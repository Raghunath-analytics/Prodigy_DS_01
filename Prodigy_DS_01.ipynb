Create a bar chart or histogram chart to visualize the distribution of a categorical or continuous variable, such as the distribution of ages or genders in a population.

```python
import pandas as pd

# Load the dataset
df = pd.read_csv('API_SP.POP.TOTL_DS2_en_csv_v2_389828.csv', skiprows=4)
```

```python
# Focus on the population values for the year 2023.
# We rename columns for clarity and drop rows with missing values (NaN).
pop_2023 = df[['Country Name', '2023']].dropna()
pop_2023.columns = ['Country', 'Population_2023']
```

```python
# Use matplotlib.pyplot to create the histogram.

import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.hist(pop_2023['Population_2023'], bins=30, color='skyblue', edgecolor='black')
plt.title('Distribution of Country Populations (2023)')
plt.xlabel('Population')
plt.ylabel('Number of Countries')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
